# Simple example of how to load image from ROS .bag

```python
bag = rosbag.Bag(test_bag, 'r')

iterator = bag.read_messages(topics=TOPIC) # select message (topic, msg, time)

# Create bridge between ROS and OpenCV
bridge = CvBridge()

for topic, msg, t in iterator:
    print(topic, t.secs)
    img = np.asarray(bridge.imgmsg_to_cv2(msg, 'bgr8'))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    pim([img], ['Topic: {} Secs: {}, H:{} W:{} dtype: {}'.format(topic, t.secs,
                                                                 msg.height, msg.width, img.dtype)])
    break
bag.close()
```
